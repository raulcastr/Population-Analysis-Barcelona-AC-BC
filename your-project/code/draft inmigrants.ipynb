```python
import pandas as pd
import numpy as np
```

**Read heads of the population data sets that we are going to use**

```python
immigrants_age=pd.read_csv("../../datasets/immigrants-emigrants-by-age.csv")
immigrants_age.head()
```

```python
immigrants_sex=pd.read_csv("../../datasets/immigrants-emigrants-by-sex.csv")
immigrants_sex.head()
```

```python
population=pd.read_csv("../../datasets/population.csv")
population.head()
```

**Calculate migratory balance and add it as a new column**

```python
migratory_balance=immigrants_sex["Immigrants"]-immigrants_sex["Emigrants"]
migratory_balance
immigrants_sex["Migratory balance"]=migratory_balance
immigrants_sex.head()
```

**Migratory balance per district from 2013 to 2017**

```python
immigrants_sex2013=immigrants_sex[(immigrants_sex["Year"]==2013)]
immigrants_sex2014=immigrants_sex[(immigrants_sex["Year"]==2014)]
immigrants_sex2015=immigrants_sex[(immigrants_sex["Year"]==2015)]
immigrants_sex2016=immigrants_sex[(immigrants_sex["Year"]==2016)]
immigrants_sex2017=immigrants_sex[(immigrants_sex["Year"]==2017)]
```

```python
migration_2013=pd.DataFrame(immigrants_sex2013.groupby(["District Code","District Name","Year"])["Migratory balance"].sum())
migration_2013=migration_2013.reset_index()
#----------------
migration_2014=pd.DataFrame(immigrants_sex2014.groupby(["District Code","District Name","Year"])["Migratory balance"].sum())
migration_2014=migration_2014.reset_index()
#----------------
migration_2015=pd.DataFrame(immigrants_sex2015.groupby(["District Code","District Name","Year"])["Migratory balance"].sum())
migration_2015=migration_2015.reset_index()
#----------------
migration_2016=pd.DataFrame(immigrants_sex2016.groupby(["District Code","District Name","Year"])["Migratory balance"].sum())
migration_2016=migration_2016.reset_index()
#----------------
migration_2017=pd.DataFrame(immigrants_sex2017.groupby(["District Code","District Name","Year"])["Migratory balance"].sum())
migration_2017=migration_2017.reset_index()
```

**Immigration growth before Colau and after Colau**

```python
def immigration_growth(dataframe1,dataframe2):
    growth=dataframe2["Migratory balance"]-dataframe1["Migratory balance"]
    
    growthdata=pd.DataFrame()
    
    growthdata["District Name"]=dataframe1["District Name"]
    growthdata["Migratory balance"]=growth

    return growthdata
```

```python
growth_BC=immigration_growth(migration_2013,migration_2015)
growth_BC
    
```

```python

growth_BC["Migratory balance"].sum()
```

```python
growth_AC=immigration_growth(migration_2015,migration_2017)
growth_AC
```

```python
growth_AC["Migratory balance"].sum()
```

```python
merged=growth_BC.merge(right=growth_AC,how="inner",on="District Name")
new_names=["District name","Migratory balance BC","Migratory balance AC"]
merged.columns=new_names
merged
```

**Total number of population per district and year**

```python
list_year=[]
list_years=[2013,2014,2015,2016,2017]
for year in list_years:
    list_year.append(population[population["Year"]==year])

population_year=[]
for i in range(len(list_years)):
    population_year.append(pd.DataFrame(list_year[i].groupby(["District.Name","District.Code","Year"])["Number"].sum()).reset_index())
population_year[2]
```

**Population growth before Colau and after Colau**

```python
def population_growth(dataframe1,dataframe2):
    growth=dataframe2["Number"]-dataframe1["Number"]
    
    growthdata=pd.DataFrame()
    
    growthdata["District Name"]=dataframe1["District.Name"]
    growthdata["Number"]=growth

    return growthdata
```

```python
population_growthBC=population_growth(population_year[0],population_year[2])
population_growthBC
```

```python
population_growthBC["Number"].sum()
```

```python
population_growthAC=population_growth(population_year[2],population_year[4])
population_growthAC
```

```python
population_growthAC["Number"].sum()
```

```python
merged2=population_growthBC.merge(right=population_growthAC,how="inner",on="District Name")
new_names=["District name","Total population BC","Total population AC"]
merged2.columns=new_names
merged2
```

```python
merged3=merged2.merge(right=merged,how="inner",on="District name")
merged3
```

```python
Immigration_over_populationBC=growth_BC["Migratory balance"].div(population_growthBC["Number"])*100
porcentajeBC=Immigration_over_populationBC.apply(func=format,args=(".0f",))
merged3["Immigration over population BC"]=porcentajeBC+"%"

Immigration_over_populationAC=growth_AC["Migratory balance"].div(population_growthAC["Number"])*100
porcentajeAC=Immigration_over_populationAC.apply(func=format,args=(".0f",))
merged3["Immigration over population AC"]=porcentajeAC+"%"
merged3
```

```python
merged4=merged3.set_index("District name").transpose()
merged4
```

```python
merged4.to_csv("../../datasets/migratory_balances.csv")
```
